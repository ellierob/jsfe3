```python
cities_conn = {
    "a": {"b": 2,"c": 1,"d": 6,"e": 2,"f": 3,"g": 0,"h": 3,"i": 3,"j": 3,"k":4},
    "b": {"a": 1,"c": 2,"d": 3,"e": 2,"f": 3,"g": 3,"h": 3,"i": 3,"j": 3,"k":5},
    "c": {"b": 2,"a": 2,"d": 3,"e": 1,"f": 3,"g": 2,"h": 3,"i": 4,"j": 3,"k":1},
    "d": {"b": 1,"c": 2,"a": 3,"e": 2,"f": 3,"g": 5,"h": 3,"i": 3,"j": 3,"k":0},
    "e": {"b": 1,"c": 5,"d": 0,"a": 2,"f": 3,"g": 2,"h": 6,"i": 3,"j": 3,"k":4},
    "f": {"b": 2,"c": 2,"d": 3,"e": 7,"a": 3,"g": 6,"h": 3,"i": 2,"j": 3,"k":5},
    "g": {"b": 7,"c": 2,"d": 3,"e": 0,"f": 3,"a": 2,"h": 3,"i": 3,"j": 3,"k":4},
    "h": {"b": 1,"c": 10,"d": 3,"e": 2,"f": 3,"g": 2,"a": 3,"i": 3,"j": 3,"k":4},
    "i": {"b": 1,"c": 2,"d": 3,"e": 8,"f": 3,"g": 2,"h": 3,"a": 3,"j": 3,"k":4},
    "j": {"b": 1,"c": 2,"d": 3,"e": 8,"f": 3,"g": 2,"h": 3,"i": 3,"a": 3,"k":4},
    "k": {"b": 1,"c": 2,"d": 3,"e": 8,"f": 3,"g": 2,"h": 3,"i": 3,"a": 3,"j":9},
}
```

```python
cities_dist = {}
cities_count = len(cities_conn.keys())
max_width = int(cities_count/3) if cities_count/3 == int(cities_count/3) else int(cities_count/3)+1
for city, conns in cities_conn.items():
    cities_dist[city] = {}
    for conn_city, conn in conns.items():
        # if city not in cities_dist:
        if conn == 0:
            cities_dist[city][conn_city] = max_width-1
        else:
            # cities_dist[city][conn_city] = int(1/conn * (max_width-1))
            cities_dist[city][conn_city] = round(1/conn * (max_width-1))
        if cities_dist[city][conn_city] == 0:
            cities_dist[city][conn_city] = 1


max_width, cities_dist
```

```python
weight_cities = {}
cities_dist_ordered = []
for p,o in cities_dist.items():
    n = 0
    for x, y in o.items():
        n+= y
    if str(n) not in weight_cities:
        weight_cities[str(n)] = [p]
    else:
        weight_cities[str(n)].append(p)
        weight_cities[str(n)] = sorted(weight_cities[str(n)])
        # print()

dist_ordered = sorted(list(map(lambda a: int(a), weight_cities.keys())),reverse=False)

for n in dist_ordered:
    for x in weight_cities[str(n)]:
        cities_dist_ordered.append(x)

        
# max_dist = max(dist_ordered) - min(sorted(cities_dist[cities_dist_ordered[0]].values(),reverse=True))

# dist_ordered
# max_dist
dist_ordered,cities_dist_ordered,weight_cities
```

```python
mid_index = int((max_width-1)/2)
# mid_index
main_city = cities_dist_ordered[0]
for idx, city_ls in nms.items():
    if main_city in city_ls:
        map_matrix[idx][mid_index] = main_city
        main_idx = idx

main_city_obj = cities_dist[main_city]

main_distance_cities = {}

for city,dist in main_city_obj.items():
    if str(dist) not in main_distance_cities.keys():
        main_distance_cities[str(dist)] = [city]
    else:
        main_distance_cities[str(dist)].append(city)
        main_distance_cities[str(dist)] = sorted(main_distance_cities[str(dist)])

main_distance_cities

main_city_distances = sorted(map(lambda x: int(x), main_distance_cities.keys()))

main_city_ordered = []
for dist in main_city_distances:
    for city in main_distance_cities[str(dist)]:
        main_city_ordered.append(city)
main_city_ordered
main_city_ordered_removeable = list(main_city_ordered)
```
